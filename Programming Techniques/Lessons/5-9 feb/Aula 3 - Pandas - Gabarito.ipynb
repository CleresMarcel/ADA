**6)** Compare a média do valor de fechamento do índice entre o primeiro e o segundo trimestre do ano estudado. Em qual dos períodos verifica-se maior média?

```python
df_primeiro_trimestre = df[(df['Date'] >= '2020-01-01') & (df['Date'] < '2020-04-01')]
df_primeiro_trimestre
```

```python
media_1trim = df_primeiro_trimestre['Close'].mean()
media_1trim
```

```python
df_segundo_trimestre = df[(df['Date'] >= '2020-04-01') & (df['Date'] < '2020-07-01')]
df_segundo_trimestre
```

```python
media_2trim = df_segundo_trimestre['Close'].mean()
media_2trim
```

```python
if media_1trim > media_2trim:
    print("A média do primeiro trimestre foi maior do que o segundo trimestre")
else:
    print("A média do primeiro trimestre foi menor do que o segundo trimestre")
```
