## Matplotlib

O **Matplotlib** é uma das principais bilbliotecas para o Python no quesito de visualização de Dados!<br><br>
Com esta biblioteca e com o auxílio de dados, conseguimos criar visualizações gráficas. Vamos começar por carregar a nossa biblioteca no notebook:

```python
# Carregando a biblioteca
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Agora vamos passar por alguns tipos de gráficos mais utilizados do Matplotlib:

### Gráfico de Barras

```python
# Criando lista de alunos e notas
alunos = ['João', 'Julia', 'Maria', 'Pedro', 'Lucas', 'Flávia']

notas = [7.5, 10, 8.5, 7, 9.5, 10]
```

```python
# Criando o gráfico de barras
plt.bar(alunos, notas)

# Mostrando o gráfico
plt.show()
```

Criamos o nosso primeiro gráfico de barras, mas faltam algumas informações como título do gráfico e eixos, além disso está acontecendo um problema nos nomes dos alunos. Vamos ver como ajustar esses pontos:

```python
# Criando o gráfico
plt.bar(alunos, notas)

# Colocar um titulo
plt.title('Notas dos Alunos da Ada')

# Colocando nome nos eixos
plt.xlabel('Nome dos Alunos')
plt.ylabel('Média das Notas de Prova')

# Rotacionando o nome dos alunos
plt.xticks(rotation = 45)

plt.show()
```

Vamos colocar mais laguns toques estéticos no nosso gráfico:<br><br>

- Aumentar a fonte do título;
- Aumentar a fonte do título dos eixos;
- Aumentar a fonte do nomes dos alunos;
- Mudar a cor do gráfico, para verde por exemplo;
- Aumentar o tamanho desse gráfico!<br><br>

Vamos ver como fazer isso então:

```python
# Configurar o tamanho da figura
plt.figure(figsize = (10,6))

# Criando o gráfico
plt.bar(alunos, notas, color = 'green')

# Colocar um titulo
plt.title('Notas dos Alunos da Ada', fontsize = 15)

# Colocando nome nos eixos
plt.xlabel('Nome dos Alunos', fontsize = 12)
plt.ylabel('Média das Notas de Prova', fontsize = 12)

# Rotacionando o nome dos alunos
plt.xticks(rotation = 45, fontsize = 10)
plt.yticks(fontsize = 10)

plt.show()
```

Vamos agora mudar a cor novamente, dessa vez para azul e vamos aproveitar melhor o tamanho do gráfico no nosso notebook. Por fim vamos salvar essa figura em um arquivo png externo:

```python
# Configurar o tamanho da figura
plt.figure(figsize = (10,6))

# Criando o gráfico
plt.bar(alunos, notas, color = 'green')

# Colocar um titulo
plt.title('Notas dos Alunos da Ada', fontsize = 15)

# Colocando nome nos eixos
plt.xlabel('Nome dos Alunos', fontsize = 12)
plt.ylabel('Média das Notas de Prova', fontsize = 12)

# Rotacionando o nome dos alunos
plt.xticks(rotation = 45, fontsize = 10)
plt.yticks(fontsize = 10)

# ajustando a figura
plt.tight_layout()

# Salvando a figura
plt.savefig('barras_verde.png')

plt.show()
```

### Gráfico de setores (Famoso Gráfico de Pizza)

```python
# Grafico de Pizza para a quantidade de alunos por turma
cursos = ['Data Science', 'Data Engineer', 'ML Engineer', 'DevOps']

qtd_alunos = [520, 276, 123, 800]
```

```python
# Criando o grafico de pizza
plt.pie(qtd_alunos, labels = cursos, autopct = '%.2f%%')

# Colocando o titulo
plt.title('Número de Matrículas por curso')

# Ajustando a figura
plt.tight_layout()

plt.show()
```

### Gráfico de Linhas

```python
dados_x = [0, 1, 2, 3, 4]
dados_y = [1, 3, 5, 7, 9]
```

```python
#Ajustando a figura
plt.figure(figsize = (10, 8))

# Grafico de linha
plt.plot(dados_x, dados_y)

# Colocando titulo
plt.title('Gráfico de Linha', fontsize = 15)

# Titulo para o eixo x
plt.xlabel('Eixo X')

plt.xticks()

# Titulo para o eixo y
plt.ylabel('Eixo Y')

plt.show()
```

##   

Assim como no Gráfico de barras, podemos customizar marcadores, cor e tipos de linha para o gráfico:

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
```

```python
#Ajustando a figura
plt.figure(figsize = (10, 8))

# Grafico de linha
plt.plot(x, y, marker = 'x', color = 'red', ls = '--')

# Colocando titulo
plt.title('Parábola', fontsize = 15)

# Titulo para o eixo x
plt.xlabel('x', fontsize = 12)

# Titulo para o eixo y
plt.ylabel('y = x²', fontsize = 12)

plt.show()
```

##   

Conseguimos também colocar dois conjuntos de dados em um mesmo gráfico:

```python
x1 = [0, 1, 2, 3, 4, 5, 6, 7, 8]
y1 = [5, 10, 15, 20, 25, 30, 35, 40, 45]

x2 = [0, 1, 2, 3, 4, 5, 6, 7, 8]
y2 = [0, 1, 4, 9, 16, 25, 36, 49, 64]
```

```python
#Ajustando a figura
plt.figure(figsize = (10, 8))

# Grafico 1
plt.plot(x1, y1, marker = 'o', color = 'green', ls = '--')

# Grafico 2
plt.plot(x2, y2, marker = 'x', color = 'blue')

# Colocando titulo
plt.title('Gráficos 1 e 2', fontsize = 15)

# Titulo para o eixo x
plt.xlabel('x', fontsize = 12)

# Titulo para o eixo y
plt.ylabel('y = x²', fontsize = 12)

# Colocando uma legenda
plt.legend(['y = 5x + 5', 'y = x²'])

plt.show()
```

### Subplots

```python
dados_x = [0, 1, 2, 3, 4, 5, 6, 7, 8]

# e 4 conjuntos diferentes para a coordenada Y
dados_y1 = [5, 10, 15, 20, 25, 30, 35, 40, 45]
dados_y2 = [0, 1, 2, 4, 8, 16, 32, 64, 128]
dados_y3 = [0.0, 0.3131151331286599, 0.5947403868365128, 0.8165528078619212, 0.9562447986054508, 0.9997675844179418, 0.9427440939531498, 0.7909091597020744, 0.5595327678952049]
dados_y4 = [1.0, 0.9497151748844606, 0.8039178268116431, 0.5772707440818498, 0.29256774453111023, -0.02155869075601868, -0.333516976054375, -0.6119335757248155, -0.828808229719958]
```

```python
# Tamanho da figura
plt.figure(figsize = (12, 12))

plt.subplot(221)
plt.plot(dados_x, dados_y1, marker = 'o', color = 'blue')
plt.title('Gráfico 1')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(222)
plt.plot(dados_x, dados_y2, marker = 'x', color = 'green')
plt.title('Gráfico 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(223)
plt.plot(dados_x, dados_y3, marker = '+', color = 'red')
plt.title('Gráfico 3')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(224)
plt.plot(dados_x, dados_y4, marker = 's', color = 'purple')
plt.title('Gráfico 4')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.show()
```

##   

E também podemos juntar alguns gráficos nos subplots:

```python
dados_y5 = [0, 2, 4, 6, 4, 2, 0, 2, 4]
```

```python
# Tamanho da figura
plt.figure(figsize = (12, 12))

plt.subplot(221)
plt.plot(dados_x, dados_y1, marker = 'o', color = 'blue')
plt.title('Gráfico 1')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(222)
plt.plot(dados_x, dados_y2, marker = 'x', color = 'green')
plt.title('Gráfico 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(223)
plt.plot(dados_x, dados_y3, marker = '+', color = 'red')
plt.plot(dados_x, dados_y5, marker = 's', color = 'orange')
plt.title('Gráfico 3 e 5')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(224)
plt.plot(dados_x, dados_y4, marker = 's', color = 'purple')
plt.title('Gráfico 4')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.show()
```
