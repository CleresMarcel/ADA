```python
# Calculando a média
def media_dados(dados:list[int]) -> float:
    soma = 0
    n = len(dados)
    for i in range(n):
        soma += dados[i]
    
    media = soma / n
    
    return media
  
```

```python
def media_dados2(dados):
    soma = 0

    for dado in dados:
        soma += dado
        
    return soma / len(dados)
```

```python
def media_dados3(dados):
    '''
    Isso é um docstring
    '''
    return sum(dados) / len(dados)
```

```python
dados = [3, 4, 5, 6]
media_dados(dados), media_dados2(dados), media_dados3(dados)
```

```python
def desvio_padrao(dados):
    soma = 0
    n = len(dados)
    media = media_dados(dados)
    
    for i in range(n):
        soma += (dados[i] - media)**2
    
    varianca = soma / n
    
    desv_pad = varianca ** (1/2)
    
    return desv_pad
```

```python
desvio_padrao(dados)
```

```python
def desvio_padrao2(dados):
    soma = 0
    media = media_dados2(dados)
    
    for elem in dados:
        soma += (elem - media)**(2)
        
    varianca = soma / len(dados)
    
    return varianca ** (0.5)
```

```python
desvio_padrao2(dados)
```

```python
def desvio_padrao3(dados):
    media = media_dados3(dados)
    
    return (sum([(elem - media)**2 for elem in dados]) / len(dados) ) **(1/2)
```

```python
desvio_padrao3(dados)
    
```

```python
# *args --> número indefinido de argumentos

def soma(*args):
    somatorio = 0
    for elem in args:
        somatorio += elem
    print(somatorio)
```

```python
soma(2, 4, 5, 3)
```

```python
# 'args' versão lista
def soma_lista(lista):
    for elem in lista:
        print(elem)
```

```python
lista = [2, 4, 5, 3, 6]

soma_lista(lista)
```

```python
def produtorio(a, b, *args):
    produto = a * b
    
    for elem in args:
        produto *= elem
    
    return produto
```

```python
produtorio(-1, 2, 3, 4)
```

```python
produtorio(3)
```

```python
produtorio(2, 3)
```

```python
def produtorio2(a, b=1, *args):
    produto = a * b
    
    for elem in args:
        produto *= elem
    
    return produto
```

```python
produtorio2(3)
```

```python
# Args devem vir após argumentos posicionais
def produtorio3(*args, a, b):
    produto = a * b
    
    for elem in args:
        produto *= elem
    
    return produto
```

```python
produtorio3(2, 3)
```

```python
# Args pode vir antes de argumentos default
def produtorio4(*args, a=2, b=3):
    produto = a * b
    
    for elem in args:
        produto *= elem
    
    return produto
```

```python
def informacoes(nome, idade = 20, *notas,**professores):
    print(f' {nome} tem {idade} anos.')
    
    if len(notas) > 0:
        print('Suas notas foram: ')
    for nota in notas:
        print(f'{nota}', sep=' - ', end=' ')
    
    if len(professores) > 0:
        print('\nMeus professores foram: ')
    for disciplina, professor in professores.items():
        print(f'{disciplina} : {professor}')
```

```python
informacoes('Mateus', 20, 9, 10, 8, 10, Matemática='Alexandre', Português='Maria') 
```

```python
informacoes('Mateus', 20, **{'Matemática':'W. Alexandre', 'Português':'Maria'}) 
```
